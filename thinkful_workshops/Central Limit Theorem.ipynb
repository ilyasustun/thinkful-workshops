```python
def create_sampling(x, size=10):
    sampling = []
    for i in range(1000):
        sample = np.random.choice(x, size=size, replace=True)
        sampling.append(sample.mean())
    sampling = np.array(sampling)
    print("Original Distribution Mean is: {}, Std Dev is: {}".format(x.mean(), x.std()))
    print("Sampling Distribution Mean is: {}, Std Dev is: {}".format(sampling.mean(), sampling.std()))
    sns.distplot(sampling);
    return None
```

```python
create_sampling(x, size= 1)
```
